```python
#importing the packages
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
#Sigmoid Function
def sigmoid_func(X, weight):
    z = np.dot(X, weight)
    return 1 / (1 + np.exp(-z))
```

```python
#Loss/Cost Function
def loss_func(X, y, weights):
    h = sigmoid_func(X, weights)
    m = y.shape[0]
    loss = 1/m * (-y * np.log(h) - (1 - y) * np.log(1 - h))
    return loss
```

```python
#Gradient Descent
def gradient_descent_func(X, h, y):
    m = y.shape[0]
    gradient_descent = 1/m * np.dot(X.T, (h - y))
    return gradient_descent

#Weight Loss Function 
def update_weight_loss_func(weight, learning_rate, gradient_descent):
    return weight - learning_rate * gradient_descent
```

```python
#Class for Logistic Regression
class AMS595_LogisticRegression:
    #Fit function
    def fit(self, X, y, epochs=25, lr=0.05):        
        loss = []
        weights = np.zeros(X.shape[1])
        m = X.shape[0]
                 
        for _ in range(epochs):        
            # Gradient Descent
            y_hat = sigmoid_func(X, weights)
            gradient_descents = gradient_descent_func(X, y_hat, y)
            weights = update_weight_loss_func(weights, lr, gradient_descents)
            # appending the loss
            loss.append(loss_func(X, y, weights)) 
            
        self.weights = weights
        self.loss = loss
    
    #Predict Function
    def predict(self, X):        
        # Predicting with sigmoid function
        z = sigmoid_func(X, self.weights)
        # Returning binary result
        return np.round(z)
```

```python
#Load the data
breast_cancer_data = datasets.load_breast_cancer()
df = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
#Display the top 5 rows
df.head(5)
```

```python
 #Test
df = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
# Add a target column, and fill it with the target data
df['target'] = breast_cancer_data.target
X = breast_cancer_data.data
# store the target data
y = breast_cancer_data.target
df
```

```python
#Train and Test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#Logistic Regression
logreg = AMS595_LogisticRegression()
logreg.fit(X_train, y_train, epochs=10000, lr=0.01)
y_pred = logreg.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print('-'*65)
print('Confusion Matrix\n')
print(confusion_matrix(y_test, y_pred))
```

```python
#Logistic Regression using sklearn
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
scikit_predict = lr_model.predict(X_test)
```
